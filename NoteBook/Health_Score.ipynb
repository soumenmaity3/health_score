```python
df["bmi"] = df["weight"] / ((df["height"] / 100) ** 2)
```

```python
df["pulse_pressure"] = df.systolic_bp - df.diastolic_bp
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('cardio',axis=1)
y = df.cardio

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scl_col = ['age','height','weight','systolic_bp','diastolic_bp','bmi','pulse_pressure']

scaler = StandardScaler()

X_train[scl_col] = scaler.fit_transform(X_train[scl_col])
X_test[scl_col] = scaler.transform(X_test[scl_col])
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier

models = {
    "Logistic": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "GradientBoost": GradientBoostingClassifier()
}
```

```python
param={
    'DecisionTree':{
        'criterion':['gini', 'entropy', 'log_loss'],
       'splitter':['best', 'random'],
       'max_depth':[3,5,2,6,1]
    },
    'RandomForest': {
        'n_estimators': [100, 200],
        'max_depth': [None, 5, 10],
        'min_samples_split': [2, 10],
        'min_samples_leaf': [1, 5],
        'class_weight': [None, 'balanced']
    },
    'GradientBoost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.05, 0.1],
        'max_depth': [3, 5]
    },
    'KNN': {
        'n_neighbors': [5, 10, 20],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    },
    'Logistic': {
        "C": [0.001, 0.01, 0.1, 1, 10, 100],
        "penalty": ["l1", "l2"],
        "solver": ["liblinear", "saga"],
        "max_iter": [100, 200, 500]
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

accuracy_report = {}

for name, model in models.items():
    print(f"Tuning {name} for accuracy...")

    para = param[name]

    gs = GridSearchCV(
        estimator=model,
        param_grid=para,
        scoring='accuracy',
        cv=3,
        n_jobs=-1
    )

    gs.fit(X_train, y_train)

    best_model = gs.best_estimator_

    y_test_pred = best_model.predict(X_test)
    accuracy = accuracy_score(y_test, y_test_pred)

    accuracy_report[name] = {
        "best_estimator": best_model,
        "accuracy": accuracy
    }
    print(f"{name} - Best Accuracy: {accuracy:.4f}")

print("\n--- Accuracy Report ---")
for model_name, scores in accuracy_report.items():
    print(f"Model: {model_name}")
    print(f"  Accuracy: {scores['accuracy']:.4f}")
```

```python
best_accuracy = -1
best_accuracy_model_name = ''

for model_name, scores in accuracy_report.items():
    if scores['accuracy'] > best_accuracy:
        best_accuracy = scores['accuracy']
        best_accuracy_model_name = model_name

best_accuracy_model = accuracy_report[best_accuracy_model_name]['best_estimator']

print(f"\nBest Model by Accuracy: {best_accuracy_model_name} with Accuracy: {best_accuracy:.4f}")
```
